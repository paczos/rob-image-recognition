## 1 Eliminacja wartości odstających

Usuwam ze zbioru uczącego próbki 642 (wartości skrajnie małe dla większości klas) oraz 186 (wartości skrajnie duże dla większości klas). Sensowność usinięcia tych dwóch próbek widoczna jest skali wykresów generowanych przez plot2features. Przed eliminacją dane tworzyły dwa zgrupowania – punktów odstających i reszty. Po usunięciu wartości odstających skala wykresów pozwala zauważyć to jak wyglądają różnice między klasami.


```octave
for i=1:size(Combs)
    ttemp = train;
    testtemp = test;

    first_idx = Combs(i, 1);
    second_idx = Combs(i, 2);

    ttemp = ttemp(:, [1 first_idx second_idx]);
    testtemp = testtemp(:, [1 first_idx second_idx]);

    % to nie jest najros�dniejszy wyb�r; 4 i 6 na pewno trzeba zmieni�
    % tutaj jawnie tworz� struktur� z parametrami dla klasyfikatora Bayesa
    % (po prawdzie, to dla funkcji licz�cej g�sto�� prawdobie�stwa) z za�o�eniem,
    % �e cechy s� niezale�ne

    pdfindep_para = para_indep(ttemp);
    pdfmulti_para = para_multi(ttemp);
    pdfparzen_para = para_parzen(ttemp, 0.001);
    % w sprawozdaniu trzeba podawa� szeroko�� okna (nie liczymy tego parametru z danych!)

    % wyniki do punktu 3
    base_ercf = zeros(1,3);
    base_ercf(1) = mean(bayescls(testtemp(:,2:end), @pdf_indep, pdfindep_para) != testtemp(:,1));
    base_ercf(2) = mean(bayescls(testtemp(:,2:end), @pdf_multi, pdfmulti_para) != testtemp(:,1));
    base_ercf(3) = mean(bayescls(testtemp(:,2:end), @pdf_parzen, pdfparzen_para) != testtemp(:,1));
    disp(Combs(i, :))
    disp(base_ercf)
end
```
